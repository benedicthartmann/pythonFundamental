```python
# Define the initial inventory as an empty list
inventory = []

# Function to list all students in the inventory
def list_inventory():
    if inventory:
        print("\nInventory list:")
        for student in inventory:
            print(f"Name: {student['name']}, Age: {student['age']}, GPA: {student['gpa']}, Hobby: {student['hobby']}")
    else:
        print("\nThe inventory is currently empty.")

# Function to add a new student to the inventory
def add_to_inventory():
    name = input("Enter student's name: ")
    age = int(input("Enter student's age: "))
    gpa = float(input("Enter student's GPA: "))
    hobby = input("Enter student's hobby: ")
    
    student = {
        'name': name,
        'age': age,
        'gpa': gpa,
        'hobby': hobby
    }
    
    inventory.append(student)
    print(f"\n{name} has been added to the inventory.")

# Function to search for a student in the inventory by name
def search_inventory():
    search_name = input("Enter the name of the student to search: ")
    found = False
    
    for student in inventory:
        if student['name'].lower() == search_name.lower():
            print(f"\nFound: Name: {student['name']}, Age: {student['age']}, GPA: {student['gpa']}, Hobby: {student['hobby']}")
            found = True
            break
    
    if not found:
        print(f"\nStudent with the name {search_name} was not found.")

# Function to delete a student from the inventory by name
def delete_from_inventory():
    delete_name = input("Enter the name of the student to delete: ")
    found = False
    
    for student in inventory:
        if student['name'].lower() == delete_name.lower():
            inventory.remove(student)
            print(f"\n{delete_name} has been deleted from the inventory.")
            found = True
            break
    
    if not found:
        print(f"\nStudent with the name {delete_name} was not found.")

# Main program loop to provide options
def main():
    while True:
        print("\nSelect an option:")
        print("1. List all Inventory")
        print("2. Add to Inventory")
        print("3. Search Inventory")
        print("4. Delete Inventory")
        print("5. Exit")
        
        option = input("Select an option: ")
        
        if option == '1':
            list_inventory()
        elif option == '2':
            add_to_inventory()
        elif option == '3':
            search_inventory()
        elif option == '4':
            delete_from_inventory()
        elif option == '5':
            print("Exiting program.")
            break
        else:
            print("Invalid option. Please try again.")

# Run the program
main()
```
