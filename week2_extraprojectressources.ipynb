Write a program that allows users to search through movie_db using different search criteria. The program should prompt the user to enter a search term, and then search through the movie database to find movies that match the search term. The user should be able to search by title, year, director, stars, and genre.

Once the user has entered a search term, the program should display a list of all the movies that match the search term, along with their details (year, director, stars, genre, and rating)

```python
movie_db = {
"The Shawshank Redemption": {
"year": 1994,
"director": "Frank Darabont",
"stars": ["Tim Robbins", "Morgan Freeman", "Bob Gunton"],
"genre": ["Drama"],
"rating": 9.3
},
"The Godfather": {
"year": 1972,
"director": "Francis Ford Coppola",
"stars": ["Marlon Brando", "Al Pacino", "James Caan"],
"genre": ["Crime", "Drama"],
"rating": 9.2
},
"The Godfather: Part II": {
"year": 1974,
"director": "Francis Ford Coppola",
"stars": ["Al Pacino", "Robert De Niro", "Robert Duvall"],
"genre": ["Crime", "Drama"],
"rating": 9.0
},
"The Dark Knight": {
"year": 2008,
"director": "Christopher Nolan",
"stars": ["Christian Bale", "Heath Ledger", "Aaron Eckhart"],
"genre": ["Action", "Crime", "Drama"],
"rating": 9.0
},
"12 Angry Men": {
"year": 1957,
"director": "Sidney Lumet",
"stars": ["Henry Fonda", "Lee J. Cobb", "Martin Balsam"],
"genre": ["Drama"],
"rating": 8.9
},
"Schindler's List": {
"year": 1993,
"director": "Steven Spielberg",
"stars": ["Liam Neeson", "Ralph Fiennes", "Ben Kingsley"],
"genre": ["Biography", "Drama", "History"],
"rating": 8.9
},
"The Lord of the Rings: The Return of the King": {
"year": 2003,
"director": "Peter Jackson",
"stars": ["Elijah Wood", "Viggo Mortensen", "Ian McKellen"],
"genre": ["Action", "Adventure", "Drama"],
"rating": 8.9
},
"Pulp Fiction": {
"year": 1994,
"director": "Quentin Tarantino",
"stars": ["John Travolta", "Uma Thurman", "Samuel L. Jackson"],
"genre": ["Crime", "Drama"],
"rating": 8.9
},
"The Lord of the Rings: The Fellowship of the Ring": {
"year": 2001,
"director": "Peter Jackson",
"stars": ["Elijah Wood", "Ian McKellen", "Orlando Bloom"],
"genre": ["Action", "Adventure", "Drama"],
"rating": 8.8
},
"Forrest Gump": {
"year": 1994,
"director": "Robert Zemeckis",
"stars": ["Tom Hanks", "Robin Wright", "Gary Sinise"],
"genre": ["Drama", "Romance"],
"rating": 8.8
},
"Inception": {
"year": 2010,
"director": "Christopher Nolan",
"stars": ["Leonardo DiCaprio", "Joseph Gordon-Levitt", "Ellen Page"],   

"genre": ["Action", "Adventure", "Sci-Fi"],   

 "rating": 8.8}

}

# Function to search movies by a given search term and criteria
def search_movie_db(search_term, search_criteria):
    results = []
    search_term = search_term.lower()

    for title, details in movie_db.items():
        if search_criteria == "title" and search_term in title.lower():
            results.append((title, details))
        elif search_criteria == "year" and details["year"] == int(search_term):
            results.append((title, details))
        elif search_criteria == "director" and search_term in details["director"].lower():
            results.append((title, details))
        elif search_criteria == "stars" and any(search_term in star.lower() for star in details["stars"]):
            results.append((title, details))
        elif search_criteria == "genre" and any(search_term in genre.lower() for genre in details["genre"]):
            results.append((title, details))

    return results

# Main function to run the search
def run_movie_search():
    while True:
        # Get search input from the user
        search_term = input("\nEnter a search term: ")
        search_criteria = input("Enter a search criteria (title, year, director, stars, genre): ").lower()

        if search_criteria not in ["title", "year", "director", "stars", "genre"]:
            print("Invalid search criteria. Please enter a valid option.")
            continue

        # Search the movie database
        matching_movies = search_movie_db(search_term, search_criteria)

        # Display results
        if matching_movies:
            for movie, details in matching_movies:
                print(f"\n\"{movie} ({details['year']})\"")
                print(f"Directed by: {details['director']}")
                print(f"Starring: {', '.join(details['stars'])}")
                print(f"Genre: {', '.join(details['genre'])}")
                print(f"Rating: {details['rating']}")
        else:
            print("\nNo movies found matching the search criteria.")

        # Ask if the user wants to perform another search
        another_search = input("\nDo you want to perform another search? (y/n): ").lower()
        if another_search != "y":
            break

# Run the movie search program
run_movie_search()
```
